```python
data=titanic_data.drop(columns=['PassengerId','Name','Ticket'])
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=['Cabin'])
```

# Dummy values

```python
from sklearn.preprocessing import LabelEncoder
data['Sex']=LabelEncoder().fit_transform(data['Sex'])
data['Embarked']=LabelEncoder().fit_transform(data['Embarked'])
data.head()
```

# Cleaning Missing Values

```python
data.shape
```

```python
data.dropna(inplace=True)
data.shape
```

```python
 data.isnull().sum()
```

```python
ax=sns.heatmap(data.corr(),annot=True)
```

```python
print(data.corr())
```

# Split Train and Test Data

```python
x=data.drop(columns='Survived')
y=data['Survived']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_train.shape
```

```python
x_test.shape
```

# Train Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
lg=LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
sv=SVC()
gb=GaussianNB()
```

```python
lg.fit(x_train,y_train)
knn.fit(x_train,y_train)
dt.fit(x_train,y_train)
rf.fit(x_train,y_train)
sv.fit(x_train,y_train)
gb.fit(x_train,y_train)
```

```python
print('Train model accuracy score')
print('LogisticRegression: ',lg.score(x_train,y_train))
print('K-Neighbors: ' ,knn.score(x_train,y_train))
print('Decision Tree: ',dt.score(x_train,y_train))
print('Random Forest: ',rf.score(x_train,y_train))
print('Support Vector: ',sv.score(x_train,y_train))
print('Gaussian NB: ',gb.score(x_train,y_train))
```

# Model Prediction Accuracy

```python
lg_pred=lg.predict(x_test)
knn_pred=knn.predict(x_test)
dt_pred=dt.predict(x_test)
rf_pred=rf.predict(x_test)
sv_pred=sv.predict(x_test)
gb_pred=gb.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print('Prediction Accuracy')
print('LogisticRegression: ',metrics.accuracy_score(y_test,lg_pred))
print('K-Neighbors: ' ,metrics.accuracy_score(y_test,knn_pred))
print('Decision Tree: ',metrics.accuracy_score(y_test,dt_pred))
print('Random Forest: ',metrics.accuracy_score(y_test,rf_pred))
print('Support Vector: ',metrics.accuracy_score(y_test,sv_pred))
print('Gaussian NB: ',metrics.accuracy_score(y_test,gb_pred))
```

```python
print("F1 Score")
print("Logistic: ",metrics.f1_score(y_test,lg_pred))
print("K-Neighbors: ",metrics.f1_score(y_test,knn_pred))
print("Decision Tree: ",metrics.f1_score(y_test,dt_pred))
print("Random Forest: ",metrics.f1_score(y_test,rf_pred))
print("Support Vector: ",metrics.f1_score(y_test,sv_pred))
print("Gaussian NB: ",metrics.f1_score(y_test,gb_pred))
```
